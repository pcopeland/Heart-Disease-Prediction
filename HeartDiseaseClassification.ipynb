```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.read_csv('heart.csv')
data2 = pd.read_csv('heart2.csv')
```

```python
print(data1.columns)
data1
```

```python
print(data2.columns)
data2
```

```python
print(data1.nunique(axis = 0))
```

```python
print(data2.nunique(axis = 0))
```

```python
print(data1['target'].value_counts())
print(data2['HeartDisease'].value_counts())
```

```python
cat1 = []
num1 = []
for column in data1.columns:
    if len(data1[column].unique()) <= 10:
        cat1.append(column)
    else:
        num1.append(column)

print('Categorical: ', cat1)
print('Continuous: ', num1)
```

```python
cat2 = []
num2 = []
for column in data2.columns:
    if len(data2[column].unique()) <= 10:
        cat2.append(column)
    else:
        num2.append(column)

print('Categorical: ', cat2)
print('Continuous: ', num2)
```

```python
df_cat1 = data1.loc[:, cat1]
for i in cat1:
    plt.figure(figsize = (15, 10))
    sns.countplot(x = i, data = df_cat1, palette = "mako")
    plt.title(i)
```

```python
df_cat2 = data2.loc[:, cat2]
for i in cat2:
    plt.figure(figsize = (15, 10))
    sns.countplot(x = i, data = df_cat2, palette = "mako")
    plt.title(i)
```

```python
plt.figure(figsize = (15, 10))
sns.set_context('notebook', font_scale = 1)
sns.heatmap(data1.corr(), annot = True, linewidth = 2)
plt.tight_layout()
```

```python
plt.figure(figsize=(15, 10))
sns.set_context('notebook', font_scale = 1)
sns.heatmap(data2.corr(),annot=True,linewidth = 2)
plt.tight_layout()
```
