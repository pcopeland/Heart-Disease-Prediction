```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_test, y_test):

    pred = clf.predict(X_test)
    clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict = True))
    cm = confusion_matrix(y_test, pred)

    TP = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TN = cm[1][1]

    print("Test Result:\n================================================")        
    print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
    print(f"False Negative Rate: {(FN / (FN + TP)) * 100:.2f}%")
    print("_______________________________________________")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")
    print("_______________________________________________")
    print(f"Confusion Matrix: \n {cm}\n")
```

## Modeling

#### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

def knn(X_train, X_test, y_train, y_test):
    knn_clf = KNeighborsClassifier()
    knn_clf.fit(X_train, y_train)

    print_score(knn_clf, X_test, y_test)

# knn(X_train, X_test, y_train, y_test)
knn(XU_train, XU_test, yU_train, yU_test)
# knn(XO_train, XO_test, yO_train, yO_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

def logR(X_train, X_test, y_train, y_test):
    lr_clf = LogisticRegression(solver = 'liblinear', random_state = 1)
    lr_clf.fit(X_train, y_train)

    print_score(lr_clf, X_test, y_test)

logR(X_train, X_test, y_train, y_test)
logR(XU_train, XU_test, yU_train, yU_test)
logR(XO_train, XO_test, yO_train, yO_test)
```

```python
from sklearn.naive_bayes import GaussianNB

def nbayes(X_train, X_test, y_train, y_test):
    nb_clf = GaussianNB()
    nb_clf.fit(X_train, y_train)

    print_score(nb_clf, X_test, y_test)

nbayes(X_train, X_test, y_train, y_test)
nbayes(XU_train, XU_test, yU_train, yU_test)
nbayes(XO_train, XO_test, yO_train, yO_test)
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

sfk = StratifiedKFold(n_splits = 10)
lr_clf = LogisticRegression(solver = 'liblinear', random_state = 1)

X2 = df2
y2 = df2.SkinCancer

def training(train, test, fold_no):

    x_train = train.drop('SkinCancer', axis = 1)
    y_train = train.SkinCancer

    x_test = test.drop('SkinCancer', axis = 1)
    y_test = test.SkinCancer

    lr_clf.fit(x_train, y_train)
    score = lr_clf.score(x_test, y_test)

    print('For Fold {} the accuracy is {}'.format(str(fold_no), score))
    
fold_no = 1

for train_index, test_index in sfk.split(X2, y2):
    train = df2.iloc[train_index, :]
    test = df2.iloc[test_index, :]
    training(train, test, fold_no)
    fold_no += 1
```
