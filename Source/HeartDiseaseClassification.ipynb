## At this point, I decided to discard data set one, since data set two is more than 300 times larger and the two sets don't share any common features. 

```python
df = data2
df.info()
```

```python
categorical_dqr = {'Feature': [], 'Desc.':[], 'Count':[], '% of Missing':[], 'Card.':[], 'Mode':[], 'Mode Freq.':[], 'Mode %':[], '2nd Mode':[], '2nd Mode Freq.':[], '2nd Mode %':[], 'Notes':[]}

cat_descriptions = ['Ever reported coronary heart disease (CHD) or myocardial infarction (MI)', 'Smoked at least 100 cigarettes in your life? [Note: 5 packs = 100 cigarettes]', 'Adult men having more than 14 drinks per week and adult women having more than 7 drinks per week', 
                    '(Ever told) (you had) a stroke?', 'Serious difficulty walking or climbing stairs?', 'Male or female?', 'Age category', 'Imputed race/ethnicity value', '(Ever told) (you had) diabetes?', 'Reported physical activity/exercise in the past 30 days, not job', 
                    'Would you say that in general your health is...', '(Ever told) (you had) asthma?', 'Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease?', '(Ever told) (you had) skin cancer?']

for i in range(len(cat2)):

    col = df[cat2[i]]
    
    if col.dtype != 'object':
        col = col.astype('object')
      
    col_vals = col.describe()
    counts = col.value_counts()
    counts_index = counts.index
     
    #Feature
    categorical_dqr['Feature'].append(cat2[i])
    #Descriptions
    categorical_dqr['Desc.'].append(cat_descriptions[i])
    #Count
    categorical_dqr['Count'].append(col_vals[0])
    #Percentage of missing values
    categorical_dqr['% of Missing'].append((col.isnull().sum() / col_vals[0])*100)
    #Cardinality
    categorical_dqr['Card.'].append(col.nunique())
    #Mode
    categorical_dqr['Mode'].append(col_vals[2])
    #Mode Freq.
    categorical_dqr['Mode Freq.'].append(col_vals[3])
    #Mode %
    categorical_dqr['Mode %'].append((col_vals[3]/col_vals[0])*100)

    #2nd Mode
    if len(counts_index) > 1:
        categorical_dqr['2nd Mode'].append(counts_index[1])
        #2nd Mode Freq.
        categorical_dqr['2nd Mode Freq.'].append(counts[1])
        #2nd Mode %
        categorical_dqr['2nd Mode %'].append((counts[1] / col_vals[0])*100)
    else:
        categorical_dqr['2nd Mode'].append('Cardinality of 1')
        categorical_dqr['2nd Mode Freq.'].append('Cardinality of 1')
        categorical_dqr['2nd Mode %'].append('Cardinality of 1')
    
    categorical_dqr['Notes'].append('')

dfqr_cat = pd.DataFrame(categorical_dqr)

dfqr_cat
```

```python
dqr_num = {'Feature': [], 'Desc.':[], 'Count':[], '% of Missing':[], 'Card.':[], 'Min.':[], 'Q1':[], 'Median':[], 'Q3':[], 'Max.':[], 'Mean':[], 'Std. Dev.':[], 'Notes':[]}

num_descriptions = ['Body Mass Index (BMI)', 'For how many days during the past 30 was your physical health not good? (0-30 days)', 'For how many days during the past 30 days was your mental health not good? (0-30 days)', 
                    'On average, how many hours of sleep do you get in a 24-hour period?']

for i in range(len(num2)):
    col = df[num2[i]]
    col_vals = col.describe()

    #print('{}\n{}\n'.format(num2[i], num2_name[i]))

    #Feature
    dqr_num['Feature'].append(num2[i])
    
    #Descriptions
    dqr_num['Desc.'].append(num_descriptions[i])
    
    #Count
    dqr_num['Count'].append(col_vals[0])
   
    #Percentage of missing values
    dqr_num['% of Missing'].append(col.isnull().sum() / col_vals[0])
   
    #Cardinality
    dqr_num['Card.'].append(col.nunique())
   
    #Min
    dqr_num['Min.'].append(col_vals[3])
   
    #q1
    dqr_num['Q1'].append(col_vals[4])
   
    #median
    dqr_num['Median'].append(col_vals[5])
   
    #q3
    dqr_num['Q3'].append(col_vals[6])
   
    #max
    dqr_num['Max.'].append(col_vals[7])
   
    #mean
    dqr_num['Mean'].append(col_vals[1])
    
    #STD
    dqr_num['Std. Dev.'].append(col_vals[2])
    
    #Notes
    dqr_num['Notes'].append('')

dfqr_num = pd.DataFrame(dqr_num)

dfqr_num
```

```python
df_drop = df.copy()
df_drop = df_drop.drop_duplicates(ignore_index = True)
drop_percentage = (len(df_drop) / len(df)) * 100

print("Percentage of values left after dropping duplicates: {}".format(drop_percentage))
df_drop.info()
df_drop
```
